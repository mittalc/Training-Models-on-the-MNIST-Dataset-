# imports

```python
# baseline cnn model for mnist
from numpy import mean
from numpy import std
from matplotlib import pyplot as plt
from sklearn.model_selection import KFold
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.optimizers import SGD, Adamax
```

# models

```python
class Model1(Sequential):	
	def __init__(self):
		super(Model1,self).__init__()
		self.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', input_shape=(28, 28, 1)))
		self.add(MaxPooling2D((2, 2)))
		self.add(Flatten())
		self.add(Dense(100, activation='relu', kernel_initializer='he_uniform'))
		self.add(Dense(10, activation='softmax'))

	def compileModel(self):
		opt = SGD(learning_rate=0.01, momentum=0.9)
		self.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
class Model2(Sequential):
    def __init__(self):
        super(Model2, self).__init__()
        self.add(Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_uniform', input_shape=(28, 28, 1)))
        self.add(BatchNormalization())
        self.add(Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_uniform', input_shape=(28, 28, 1)))
        self.add(BatchNormalization())
        self.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', input_shape=(28, 28, 1)))
        self.add(MaxPooling2D((2, 2)))
        self.add(Flatten())
        self.add(Dense(100, activation='relu', kernel_initializer='he_uniform'))
        self.add(Dense(10, activation='softmax'))
        
    def compileModel(self):
        # compile self
        opt = SGD(learning_rate=0.01, momentum=0.9)
        self.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
class Model3(Sequential):
    def __init__(self):
        super(Model3,self).__init__()
        self.add(Conv2D(128, (3,3), activation='relu', kernel_initializer='he_uniform', input_shape=(28, 28, 1)))
        self.add(BatchNormalization())

        self.add(Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_uniform', input_shape=(28, 28, 1)))
        self.add(BatchNormalization())

        self.add(Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_uniform', input_shape=(28, 28, 1)))
        self.add(BatchNormalization())

        self.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', input_shape=(28, 28, 1)))
        self.add(BatchNormalization())
        
        self.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', input_shape=(28, 28, 1)))
        self.add(MaxPooling2D((2, 2)))
        self.add(BatchNormalization())

        self.add(Flatten())
        self.add(Dense(128, activation='relu', kernel_initializer='he_uniform'))
        self.add(BatchNormalization())

        self.add(Dropout(0.2))
        self.add(Dense(64, activation='relu', kernel_initializer='he_uniform')) 
        self.add(BatchNormalization())

        self.add(Dropout(0.2))
        self.add(Dense(10, activation='softmax'))
        
    def compileModel(self):
        # compile self
        opt = Adamax(learning_rate = 0.00075)
        self.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
```

# main

```python

# load train and test dataset
def load_dataset():
	# load dataset
	(trainX, trainY), (testX, testY) = mnist.load_data()
	# reshape dataset to have a single channel
	trainX = trainX.reshape((trainX.shape[0], 28, 28, 1))
	testX = testX.reshape((testX.shape[0], 28, 28, 1))
	# one hot encode target values
	trainY = to_categorical(trainY)
	testY = to_categorical(testY)
	return trainX, trainY, testX, testY

# scale pixels
def prep_pixels(train, test):
	# convert from integers to floats
	train_norm = train.astype('float32')
	test_norm = test.astype('float32')
	# normalize to range 0-1
	train_norm = train_norm / 255.0
	test_norm = test_norm / 255.0
	# return normalized images
	return train_norm, test_norm

# define cnn model
def define_model(m):
	# Loads the corresponding model
	if m == 'm1':
		model = Model1()
	elif m == 'm2':
		model = Model2()
	elif m == 'm3':
		model = Model3()
	model.compileModel()
	return model

# evaluate a model using k-fold cross-validation
def evaluate_model(dataX, dataY, testX, testY, EPOCHS, BATCHSIZE, MODEL, n_folds=5):
	scores, histories = list(), list()
 	# If n_folds is 1, then don't use sklearn and just manual seperate the dataset into a train and validate set
	if n_folds == 1:
		trainX, trainY, valX, valY = dataX[:50000], dataY[:50000], dataX[50000:], dataY[50000:]
		model = define_model(MODEL)
		# fit model
		history = model.fit(trainX, trainY, epochs=EPOCHS, batch_size=BATCHSIZE, validation_data=(valX, valY), verbose=2)
		# evaluate model on the test values given by the MNIST dataset
		_, acc = model.evaluate(testX, testY, verbose=0)
		print('> %.3f' % (acc * 100.0))
		# stores scores
		scores.append(acc)
		histories.append(history)
	else:
		# prepare cross validation
		kfold = KFold(2, shuffle=True, random_state=1)
		# enumerate splits
		for train_ix, test_ix in kfold.split(dataX):
			# define model
			model = define_model()
			# select rows for train and test
			trainX, trainY, valX, valY = dataX[train_ix], dataY[train_ix], dataX[test_ix], dataY[test_ix]
			# fit model
			history = model.fit(trainX, trainY, epochs=EPOCHS, batch_size=BATCHSIZE, validation_data=(valX, valY), verbose=1)
			# evaluate model on the test values given by the MNIST dataset
			_, acc = model.evaluate(testX, testY, verbose=0)
			print('> %.3f' % (acc * 100.0))
			# stores scores
			scores.append(acc)
			histories.append(history)
	return scores, histories

# plot diagnostic learning curves
def summarize_diagnostics(histories, e, bz ,m):
	print(" summarizing_training")
	for i in range(len(histories)):
		# plot loss
		plt.subplot(2, 1, 1)
		plt.title('Cross Entropy Loss')
		plt.plot(histories[i].history['loss'], color='blue', label='train')
		plt.plot(histories[i].history['val_loss'], color='orange', label='test')
		# plot accuracy
		plt.subplot(2, 1, 2)
		plt.title('Classification Accuracy') 
		plt.plot(histories[i].history['accuracy'], color='blue', label='train')
		plt.plot(histories[i].history['val_accuracy'], color='orange', label='test')
	 	# change added - Chirag to view
		plt.tight_layout()
		# saves the plot of the training
		filename = 'Train-Model,'+str(m)+';Epochs,'+str(e)+';BatchSize,'+str(bz)+'.png'
		plt.savefig(filename, format='png')
		plt.show()
		# Clear plot so graphs don't overlap
		plt.clf()

# summarize model performance
def summarize_performance(scores, e, bz ,m):
	print("summarizing_performance")
	# print summary
	print('Accuracy: mean=%.3f std=%.3f, n=%d' % (mean(scores)*100, std(scores)*100, len(scores)))
	# box and whisker plots of results
	plt.boxplot(scores)
	# Saves the plot of the testing
 	# filename = 'Plots\\Evaluate-Model,'+str(m)+';Epochs,'+str(e)+';BatchSize,'+str(bz)+'.png'
	# plt.savefig(filename,format='png')
	plt.show()
	#print(" saving " + "epoch" + str(e)  +" batchsize" + str(bz) + '.png' )
	

# run the test harness for evaluating a model
def run_test_harness():
	# load dataset
	trainX, trainY, testX, testY = load_dataset()
	# prepare pixel data
	trainX, testX = prep_pixels(trainX, testX)
	# evaluate model
	for e in EPOCHS:
		for bz in BATCHSIZE:
			for m in MODELS:
				print('Training Model:',m," Epochs:", str(e), " BatchSize:", str(bz))
				scores, histories = evaluate_model(trainX, trainY, testX, testY, e, bz, m, n_folds=1)
				# learning curves
				summarize_diagnostics(histories,e, bz, m)
				# summarize estimated performance
				summarize_performance(scores, e, bz, m)

# entry point, run the test harness

EPOCHS = [2,5,10]
BATCHSIZE  = [12,16,24,32]
MODELS = ['m1','m2','m3']

run_test_harness()
```
